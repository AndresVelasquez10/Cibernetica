```python
import random
import time

class AgenteAutonomo:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.direccion = None

    def mover(self):
        direcciones = ['arriba', 'abajo', 'izquierda', 'derecha']
        self.direccion = random.choice(direcciones)

        if self.direccion == 'arriba':
            self.y += 1
        elif self.direccion == 'abajo':
            self.y -= 1
        elif self.direccion == 'izquierda':
            self.x -= 1
        elif self.direccion == 'derecha':
            self.x += 1

    def mostrar_posicion(self):
        print(f"Posición actual: ({self.x}, {self.y})")
        return (self.x,self.y)

    def actuar(self, tiempo=10000):
        mov = []
        inicio = time.time()
        while time.time() - inicio < tiempo:
            self.mover()
            mov.append(self.mostrar_posicion())
            time.sleep(1)
        return mov

agente = AgenteAutonomo(0, 0)

mov = agente.actuar(tiempo=30)
```

```python
import matplotlib.pyplot as plt

x, y = zip(*mov)

plt.figure(figsize=(6, 6))
for i in range(len(x)):
    plt.plot(x[:i], y[:i], marker='o', linestyle='-', color='b')
    plt.scatter(x[i], y[i], color='r', label='Estado actual' if i == len(x)-1 else "")
    plt.pause(0.5)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Movimiento del objeto paso a paso")
plt.grid(True)
plt.show()
```
